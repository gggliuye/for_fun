```python
import numpy as np
import matplotlib.pyplot as plt
```

## Love Affairs problem

$$
\dot{R} = aR + bJ
$$

$$
\dot{J} = cR + dJ
$$

```python
def dev_affairs(x, a_mat):
    """ the gradient of the love affairs. We are dealing 2d problem here.
    Hence the input states must be 2d vectors.
    we have dot(x) = A * x
    
    Args:
        x (2*n array) : for the current state
        a_mat (2*2 array) : for the state matrix
    """
    return np.dot(a_mat, x)
```

```python
def plot_derivative(a_mat, range_m = 8, reso = 20):
    # create initial values for the grid plot
    rs = np.linspace(-range_m, range_m, reso)
    js = np.linspace(-range_m, range_m, reso)

    r_grid, j_grid = np.meshgrid(rs, js)

    # make into array for fast process
    r_array = r_grid.reshape([1, -1])
    j_array = j_grid.reshape([1, -1])
    r_j_array = np.concatenate((r_array, j_array), axis=0)

    deri_array = dev_affairs(r_j_array, a_mat)
    
    # reshape back intogrid
    us = deri_array[0, :]
    vs = deri_array[1, :]
    
    fig = plt.quiver(r_array, j_array, us, vs, color='r')
```

## Out of touch with their own feelings

$$
\begin{bmatrix} 0 & a \\ b & 0 \end{bmatrix}
$$


```python
plt.figure(figsize=(20, 5))
plt.subplot(1,4,1)
plot_derivative(np.array([[0, 1], [1, 0]]))
tit = plt.title("a > 0, b > 0")

plt.subplot(1,4,2)
plot_derivative(np.array([[0, -1], [-1, 0]]))
tit = plt.title("a < 0, b < 0")

plt.subplot(1,4,3)
plot_derivative(np.array([[0, 1], [-1, 0]]))
tit = plt.title("a > 0, b < 0")

plt.subplot(1,4,4)
plot_derivative(np.array([[0, 1], [-1, 0]]))
tit = plt.title("a < 0, b > 0")
```

## Fire and water 
do opposite attract ?

$$
\begin{bmatrix} a & b \\ -b & -a \end{bmatrix}
$$


```python
plt.figure(figsize=(20, 5))
plt.subplot(1,4,1)
plot_derivative(np.array([[1, 2], [-2, -1]]))
tit = plt.title("|b| > |a|")

plt.subplot(1,4,2)
plot_derivative(np.array([[1, 1], [-1, -1]]))
tit = plt.title("|b| = |a|")

plt.subplot(1,4,3)
plot_derivative(np.array([[3, 1], [-1, -3]]))
tit = plt.title("|b| < |a|, a > 0, b > 0")

plt.subplot(1,4,4)
plot_derivative(np.array([[3, -1], [1, -3]]))
tit = plt.title("|b| < |a|, a > 0, b < 0")
```

## Romantic clones

$$
\begin{bmatrix} a & b \\ b & a \end{bmatrix}
$$


```python
plt.figure(figsize=(20, 5))
plt.subplot(1,5,1)
plot_derivative(np.array([[1, 2], [2, 1]]))
tit = plt.title("a > 0, b > 0")

plt.subplot(1,5,2)
plot_derivative(np.array([[1, -2], [-2, 1]]))
tit = plt.title("a > 0, b < 0, |a| < |b|")

plt.subplot(1,5,3)
plot_derivative(np.array([[1, -0.5], [-0.5, 1]]))
tit = plt.title("a > 0, b < 0, |a| > |b|")

plt.subplot(1,5,4)
plot_derivative(np.array([[-1, 2], [2, -1]]))
tit = plt.title("a < 0, b > 0, |a| < |b|")

plt.subplot(1,5,5)
plot_derivative(np.array([[-1, 0.1], [0.1, -1]]))
tit = plt.title("a < 0, b > 0, |a| > |b|")
```
